# Get coins in each categories

```python
display_currency = 'usd'
# category = 'stablecoins' # use only for individual categories
display_order = 'market_cap_desc'
sparkline = 'false'
```

```python
raw_category_list_df = ['aave-tokens','analytics','arbitrum-ecosystem']
```

```python
coin_in_category_df = pd.DataFrame()
for category in raw_category_list_df:
    endpoint_coin_in_category = f'coins/markets?vs_currency={display_currency}&category={category}&order={display_order}&sparkline={sparkline}'
    request_coin_in_category = requests.get(main_url + endpoint_coin_in_category)
    temp_coin_in_category_df = pd.read_json(request_coin_in_category.text)
    temp_coin_in_category_df = temp_coin_in_category_df.sort_values('market_cap',ascending=False).head(30)
    temp_coin_in_category_df['category'] = category
    coin_in_category_df = pd.concat([coin_in_category_df, temp_coin_in_category_df],axis=0)
    
```
